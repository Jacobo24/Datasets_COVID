```python
columnas_drop = ["user_name", "user_location", "user_description", "user_created", "date", "hashtags", "source", "text"]
```

```python
df_covid2 = df_covid.drop(columns = columnas_drop, axis=1)
print("Tama√±o del dataset original: ", df_covid.shape)
print("Tama√±o del dataset nuevo: ", df_covid2.shape)
```

```python
df_covid2.dtypes.value_counts()
```

```python
target = "user_verified"
features = [x for x in df_covid2.columns if x != target]

print ("Features: ", features)
print ("Target: ", target)
```

```python
df_covid2[target]
```

```python
from sklearn.tree import DecisionTreeRegressor

arbol_importancia = DecisionTreeRegressor(max_depth=len(features)+10, random_state=42)

arbol_importancia.fit(X=df_covid2[features], y=df_covid2[target])
```

```python
y_pred_arbol = arbol_importancia.predict(X=df_covid2[features])

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

print('Mean Absolute Error:', mean_absolute_error(df_covid2[target], y_pred_arbol))
print('Mean Absolute Percentage Error:', mean_absolute_percentage_error(df_covid2[target], y_pred_arbol)*100)
print('Mean Squared Error:', mean_squared_error(df_covid2[target], y_pred_arbol))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(df_covid2[target], y_pred_arbol)))
print('R^2 coefficient of determination:', r2_score(df_covid2[target], y_pred_arbol))
```
